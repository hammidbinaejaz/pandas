# series in pandas

```python
import numpy as np
import pandas as pd
```

# series from lists

# string 

```python
country = ["india", 'pakistan','usa','nepal','srilanka']
pd.Series(country)
```

```python
# integers 
runs = [13,22,33,44,55,56,677]
pd.Series(runs)
```

# custom index.
marks = [9,8,7,6,5]
sub = ['maths','phy','chem','bio','eng']

pd.Series(marks,index=sub)

```python
# setting a name
```

```python
marks = pd.Series(marks,index=sub,name='hammid k marks')
```

# series from dicts

```python
marks = {
    'maths':9,
    'phy' :8,
    'chem':7,
    'bio':6,
    'eng':5
}
markd_series = pd.Series(marks,name="hello")
markd_series
```
